```python
#
import numpy as np
import matplotlib.pyplot as plt
#
plt.rcParams['font.sans-serif']=['Arial']    #Arial
plt.rcParams['axes.unicode_minus']=False   #
#
def lorentzian(x, x0, gamma, A):
    """
    
    x:
    x0:
    gamma:(FWHM)
    A:
    """
    return A*(gamma**2/((x-x0)**2+gamma**2))
#
x = np.linspace(1.4, 3, 1000)
y1 = lorentzian(x,1.7, 0.003, 1)
y2 = lorentzian(x, 2, 0.1, 0.8)
y3 = lorentzian(x,2.4,0.3,0.6)
y_total = y1+y2+y3
plt.figure(figsize=(10,8))
plt.plot(x, y1, 'b--', alpha=0.7, linewidth=2, label='P1 (x0=1.7, =0.03, A=1)')
plt.plot(x, y2, 'g--', alpha=0.7, linewidth=2, label='P2 (x0=2, =0.1, A=0.8)')
plt.plot(x, y3, 'r--', alpha=0.7, linewidth=2, label='P3 (x0=2.4, =0.5, A=0.6)')
plt.plot(x, y_total, 'k-', linewidth=2, label='Total')

#
plt.xlabel('Energy (eV)', fontsize=16)   #x
plt.ylabel('Intensity (a.u.)', fontsize=16)   #y
plt.tick_params(axis='both', which='major', labelsize=14)   #

plt.title('Lorentzian PL', fontsize=18)   #
plt.legend(fontsize=16, frameon=False)                   #

# 
ax = plt.gca()  # 
ax.spines['top'].set_linewidth(1.5)     # 
ax.spines['right'].set_linewidth(1.5)   # 
ax.spines['bottom'].set_linewidth(1.5)  # 
ax.spines['left'].set_linewidth(1.5)    # 

plt.tight_layout()
plt.show()
```
