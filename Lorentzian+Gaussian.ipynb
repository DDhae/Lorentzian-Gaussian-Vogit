```python
#
import numpy as np
import matplotlib.pyplot as plt
#
plt.rcParams['font.sans-serif']=['Arial']    #
plt.rcParams['axes.unicode_minus']=False   #
#
def lorentzian(x, x0, gamma, A):
    """
    
    x:
    x0:
    gamma:(FWHM)
    A:
    """
    return A*(gamma**2/((x-x0)**2+gamma**2))

def gaussian(x, x0, sigma, A):
    """
    
    x: 
    A: 
    x0: 
    sigma: 
     FWHM = 2 * sqrt(2 * ln(2)) * sigma  2.355 * sigma
    """
    return A * np.exp(-(x - x0)**2 / (2 * sigma**2))
    
#
x = np.linspace(1.24, 3.6, 1000)
y1 = lorentzian(x,1.7, 0.003, 1)
y2 = lorentzian(x, 2, 0.08, 0.8)
y3 = lorentzian(x,2.4,0.12,0.6)
y4 = gaussian(x, 1.4, 0.08, 0.3)
y5 = gaussian(x,2.6,0.12,0.8)

y_total = y1+y2+y3+y4+y5
plt.figure(figsize=(10,8))
plt.plot(x, y1, 'r--', alpha=0.7, linewidth=2, label='L1 (x0=1.7, =0.03, A=1)')
plt.plot(x, y2, 'm--', alpha=0.7, linewidth=2, label='L2 (x0=2, =0.08, A=0.8)')
plt.plot(x, y3, 'y--', alpha=0.7, linewidth=2, label='L3 (x0=2.4, =0.12, A=0.6)')
plt.plot(x, y4, 'b--', alpha=0.7, linewidth=2, label='G2 (x0=1.4, =0.08, A=0.3)')
plt.plot(x, y5, 'g--', alpha=0.7, linewidth=2, label='G3 (x0=2.6, =0.12, A=0.8)')

plt.plot(x, y_total, 'k-', linewidth=2, label='Total')

#
plt.xlabel('Energy (eV)', fontsize=16)   #x
plt.ylabel('Intensity (a.u.)', fontsize=16)   #y
plt.tick_params(axis='both', which='major', labelsize=14)   #

plt.title('PL', fontsize=18)   #
plt.legend(fontsize=16, frameon=False)                   #

# 
ax = plt.gca()  # 
ax.spines['top'].set_linewidth(1.5)     # 
ax.spines['right'].set_linewidth(1.5)   # 
ax.spines['bottom'].set_linewidth(1.5)  # 
ax.spines['left'].set_linewidth(1.5)    # 

plt.tight_layout()
plt.show()
```
