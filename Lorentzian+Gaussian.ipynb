```python
#设置环境
import numpy as np
import matplotlib.pyplot as plt
#设置中文字体
plt.rcParams['font.sans-serif']=['Arial']    #使用黑体
plt.rcParams['axes.unicode_minus']=False   #正确使用负号
#定义洛伦兹函数
def lorentzian(x, x0, gamma, A):
    """
    洛伦兹函数
    x:自变量
    x0:中心位置
    gamma:半高全宽(FWHM)
    A:峰值幅度
    """
    return A*(gamma**2/((x-x0)**2+gamma**2))

def gaussian(x, x0, sigma, A):
    """
    高斯函数公式
    x: 自变量
    A: 峰高
    x0: 峰中心位置
    sigma: 标准差
    注意：半高全宽 FWHM = 2 * sqrt(2 * ln(2)) * sigma ≈ 2.355 * sigma
    """
    return A * np.exp(-(x - x0)**2 / (2 * sigma**2))
    
#生成数据，或者说是确定参数
x = np.linspace(1.24, 3.6, 1000)
y1 = lorentzian(x,1.7, 0.003, 1)
y2 = lorentzian(x, 2, 0.08, 0.8)
y3 = lorentzian(x,2.4,0.12,0.6)
y4 = gaussian(x, 1.4, 0.08, 0.3)
y5 = gaussian(x,2.6,0.12,0.8)

y_total = y1+y2+y3+y4+y5
plt.figure(figsize=(10,8))
plt.plot(x, y1, 'r--', alpha=0.7, linewidth=2, label='L1 (x0=1.7, γ=0.03, A=1)')
plt.plot(x, y2, 'm--', alpha=0.7, linewidth=2, label='L2 (x0=2, γ=0.08, A=0.8)')
plt.plot(x, y3, 'y--', alpha=0.7, linewidth=2, label='L3 (x0=2.4, γ=0.12, A=0.6)')
plt.plot(x, y4, 'b--', alpha=0.7, linewidth=2, label='G2 (x0=1.4, γ=0.08, A=0.3)')
plt.plot(x, y5, 'g--', alpha=0.7, linewidth=2, label='G3 (x0=2.6, γ=0.12, A=0.8)')

plt.plot(x, y_total, 'k-', linewidth=2, label='Total')

#设置坐标轴字体大小
plt.xlabel('Energy (eV)', fontsize=16)   #x轴标签字体大小
plt.ylabel('Intensity (a.u.)', fontsize=16)   #y轴标签字体大小
plt.tick_params(axis='both', which='major', labelsize=14)   #坐标轴刻度字体大小

plt.title('PL', fontsize=18)   #标题字体大小
plt.legend(fontsize=16, frameon=False)                   #图例字体大小，去除图例边框

# 设置坐标轴线宽
ax = plt.gca()  # 获取当前坐标轴
ax.spines['top'].set_linewidth(1.5)     # 上边框线宽
ax.spines['right'].set_linewidth(1.5)   # 右边框线宽
ax.spines['bottom'].set_linewidth(1.5)  # 下边框线宽
ax.spines['left'].set_linewidth(1.5)    # 左边框线宽

plt.tight_layout()
plt.show()
```
